```python
import pandas as pd

df = pd.read_csv('datasets/sampled_posts.csv')
df
```

# LangDetect

https://pypi.org/project/langdetect/

Considering the senarios of text consisting of only emojis:



```python
from langdetect import detect
import time

def pre_processing(text):
    
    return text

start_time = time.time()  # Record the start time
for i in range(len(df)):
    text = df.loc[i, 'text']
    # print(text)
    processed_text = pre_processing(text)
    # print(processed_text)
    try:
        temp_lan = detect(processed_text)
    except:
        # print('error occurs for the text of:', processed_text)
        temp_lan = 'ERROR'

    df.loc[i, 'lang_langdetect'] = temp_lan

end_time = time.time()  # Record the end time

execution_time = end_time - start_time
print(f"Execution time: {execution_time:.6f} seconds")
print(f"The execution time for 1 million would be around: {execution_time*1000000/len(df)/3600:.6f} hours")

df
```

```python
from langdetect import detect_langs
import time

def pre_processing(text):
    
    return text

start_time = time.time()  # Record the start time
for i in range(len(df)):
    text = df.loc[i, 'text']
    # print(text)
    processed_text = pre_processing(text)
    # print(processed_text)
    try:
        temp_lan = detect_langs(processed_text)
    except:
        # print('error occurs for the text of:', processed_text)
        temp_lan = 'ERROR'

    df.loc[i, 'lang_langdetect_prob'] = str(temp_lan)

end_time = time.time()  # Record the end time

execution_time = end_time - start_time
print(f"Execution time: {execution_time:.6f} seconds")
print(f"The execution time for 1 million would be around: {execution_time*1000000/len(df)/3600:.6f} hours")

df
```
